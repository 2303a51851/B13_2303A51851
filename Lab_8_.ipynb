```python
import seaborn as sns
import matplotlib.pyplot as plt
tips=sns.load_dataset("tips")
#Boxplot
sns.boxenplot(x="day",y="total_bill",data=tips)
#Violinplot
sns.violinplot(x="day",y="total_bill",data=tips)
plt.title("Boxplot of total bill by day")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("/content/Student_Marks.csv")

# Check first rows (optional)
print(df.head())

# Plot bar chart
plt.figure(figsize=(8, 6))
bars = plt.bar(df['number_courses'], df['Marks'], color=['skyblue', 'lightgreen', 'salmon', 'orange', 'violet'])

# Add labels and title
plt.xlabel("Number of Courses", fontsize=12)
plt.ylabel("Marks", fontsize=12)
plt.title("Marks of Students by Number of Courses", fontsize=14)
plt.xticks(rotation=0) # Removed rotation as x-axis values are numerical

# Annotate values on top of bars
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 1,
             str(round(bar.get_height(), 1)),
             ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()

# Insights: find max and min marks
highest_marks = df.loc[df['Marks'].idxmax()]
lowest_marks = df.loc[df['Marks'].idxmin()]

print(f"Highest Marks: {highest_marks['Marks']} with {highest_marks['number_courses']} courses and {highest_marks['time_study']} hours of study")
print(f"Lowest Marks: {lowest_marks['Marks']} with {lowest_marks['number_courses']} courses and {lowest_marks['time_study']} hours of study")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("/content/ElectronicsData.csv")

# Preview dataset (optional)
print(df.head())

# Convert 'Price' column to numeric, removing '$' and handling potential errors
df['Price'] = pd.to_numeric(df['Price'].replace({'\$': ''}, regex=True), errors='coerce')

# Remove rows with NaN in 'Price' column after conversion
df.dropna(subset=['Price'], inplace=True)

# Plot bar chart
plt.figure(figsize=(8, 6))
bars = plt.bar(df['Sub Category'], df['Price'], color=plt.cm.tab10.colors)

# Add labels and title
plt.xlabel("Product Sub Category", fontsize=12)
plt.ylabel("Price", fontsize=12)
plt.title("Price Distribution by Product Sub Category", fontsize=14)
plt.xticks(rotation=30)

# Annotate values on top of bars
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 1,
             str(round(bar.get_height(), 1)),
             ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()

# Insights: highest priced category
highest_price = df.loc[df['Price'].idxmax()]
print(f"Highest Priced Sub Category: {highest_price['Sub Category']} with Price = {highest_price['Price']}")
```
