```python
import matplotlib.pyplot as plt

months=["jan","feb","march","april","may","june","july","aug","sep","oct","nov","dec"]
sales=[598,345,2342,432,1432,5423,1234,14321,144,13413,123,13123]

plt.bar(months,sales, color='yellow')

plt.title("Monthly sales")
plt.xlabel("Months")
plt.ylabel("Sales in USD ")
plt.grid(True,color='green')
plt.show()
```

```python
#create for the stock price

import matplotlib.pyplot as plt
days=["mon","tues","wed","thur","fri","sat","sun"]
stock_price=[224,345,242,132,143,242,2345]
plt.plot(days,stock_price,marker='^',linestyle='--',color='red',linewidth=3)
plt.title("Stock price")
plt.xlabel("Days")
plt.ylabel(" in USD")
plt.grid(True,color='green')
plt.show()
```
